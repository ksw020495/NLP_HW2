```python
import requests
from bs4 import BeautifulSoup
import re

pages = set()
movie_link = set()
movie_num = set()

def getLinks(pageUrl):
    global pages
    url = pageUrl
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html.parser")
    
    for link in soup.findAll("a", href=re.compile("^(https://movies.yahoo.com.tw/)")):
        if 'href' in link.attrs:
            if link.attrs['href'] not in pages:
                #We have encountered a new page
                newPage = link.attrs['href']
                if newPage[:43] == "https://movies.yahoo.com.tw/movieinfo_main/":
                    print(newPage[-5:])
                    if newPage[-5:] not in movie_num:
                        movie_num.add(newPage[-5:])
                        movie_link.add(newPage)
                        print(len(movie_link))
                if len(movie_link) < 100:
                    pages.add(newPage)
                    getLinks(newPage)

getLinks("https://movies.yahoo.com.tw/") 
```

```python
import requests
from bs4 import BeautifulSoup

infos = []

for num in range(1, 5):
    url = "https://movies.yahoo.com.tw/movieinfo_main/" + str(num)
    r = requests.get(url)
    print(url)
    soup = BeautifulSoup(r.text)
    for d in soup.find_all('div', id="content_l"):
        info = {}
        
        info[''] = d.find('div', class_='movie_intro_info_r').find('h1').text.strip()
        
        movie_class = []
        for cl in d.find('div', class_='level_name_box').find_all('div', class_='level_name'):
            movie_class.append(cl.find('a').text.strip())
        info[''] = movie_class
        
        info[''] = d.find('span', id ='story').text.strip()
        info[''] = d.find('span').text.strip()[5:]
        
        infos.append(info)
        
        print(info)

infos
```
